```python
#Saving the heatmap
plt.matshow(df.corr())
plt.savefig("heatmap.png")
```

```python
#Creating a correlation heatmap using seaborn and adding labels, a legend, and change the size of the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#comments
Correlation coefficient between charge and age is 0.3 meaning a moderate positive correlation. This could be interpreted as older individuals tend to have higher medical charges compared to younger individuals.

Correlation coefficient between charge and BMI is 0.2 meaning a weak positive correlation. This could be interpreted as higher BMI is positively associated with increased medical charges.
```

```python
##Saving the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
plt.savefig("Correlation heatmap,png")
```

```python
#Scatterplots using seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='charges')
plt.title('Scatterplot of Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#comments
The scatterplot shows a positive linear relationship between age and charge
```

```python
# Creating a pair plot using Seaborn
sns.pairplot(df[['age', 'bmi', 'charges']])
plt.suptitle('Pair Plot of Age, BMI, and Charges')
plt.show()
```

```python
#comments
The scatterplot shows that there is a closer relationship between age and charges than between BMI and charges.
```

```python
# Creating a categorical plot using Seaborn catplot
sns.catplot(x='sex', y='charges', hue='smoker', kind='box', data=df)
plt.title('Categorical Plot of Charges by Sex and Smoking Status')
plt.show()
```

```python
#Comments
Male smokers pay more insurance charges compared to female smokers.
```
